```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
dataset.columns
```

```python
inde=dataset[["age","bmi","children","sex_male","smoker_yes"]]
```

```python
dep=dataset[["charges"]]
```

```python
inde
```

```python
dep
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.tree import DecisionTreeRegressor
param_grid={'criterion':['squared_error','friedman_mse','absolute_error','poisson'],'max_features':['sqrt','log2'],
            'splitter':['best','random']}
grid=GridSearchCV(DecisionTreeRegressor(),param_grid,verbose=3,refit=True,n_jobs=-1)
```

```python
grid.fit(inde,dep)
```

```python
result=grid.cv_results_
```

```python
result
```

```python
table=pd.DataFrame.from_dict(result)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children"))
sex_male_input=int(input("Sex_Male 0 or 1:"))
smoker_yes_input=int(input("Smoker yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
