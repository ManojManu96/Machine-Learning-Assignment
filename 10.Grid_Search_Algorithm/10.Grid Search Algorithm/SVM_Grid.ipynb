```python
import pandas as pd
```

```python
dataset=pd.read_csv("insurance_pre.csv")
```

```python
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True,dtype=int)
```

```python
dataset
```

```python
dataset.columns
```

```python
inde=dataset[["age","bmi","children","sex_male","smoker_yes"]]
```

```python
inde
```

```python
dep=dataset[["charges"]]
```

```python
from sklearn.model_selection import GridSearchCV
from sklearn.svm import SVR
param_grid={'kernel':['rbf','sigmoid','linear'],'C':[10,100,1000,2000,3000],'gamma':['auto','scale']}
grid=GridSearchCV(SVR(),param_grid,refit=True,verbose=3,n_jobs=-1)
grid.fit(inde,dep)
```

```python
results=grid.cv_results_
print("The R_score value for best parameter{}:".format(grid.best_params_))
```

```python
table=pd.DataFrame.from_dict(results)
```

```python
table
```

```python
age_input=float(input("Age:"))
bmi_input=float(input("BMI:"))
children_input=float(input("Children"))
sex_male_input=int(input("Sex_Male 0 or 1:"))
smoker_yes_input=int(input("Smoker yes 0 or 1:"))
```

```python
Future_Prediction=grid.predict([[age_input,bmi_input,children_input,sex_male_input,smoker_yes_input]])
print("Future_Prediction={}".format(Future_Prediction))
```
