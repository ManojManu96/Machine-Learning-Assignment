```python
# pandas used to read data from database related tables...
import pandas as pd
```

```python
# Pd is a class ,it call the csv file
dataset=pd.read_csv("Salary_Data.csv")
```

```python
dataset
```

```python
# assign column for input variable as independent
independent=dataset[["YearsOfExperience"]]
```

```python
# view what will present inside indeependent variable
independent
```

```python
# # assign column for output variable as dependent
dependent=dataset[["Salary"]]
```

```python
# view what will present inside deependent variable
dependent
```

```python
# split train_set and test_set by using sklearn 
# from sklearn folder inside haveing model_selection to call train_test_split function
# here we calling only particular train_test_split function ,it use reduced size of memory
from sklearn.model_selection import train_test_split
# call input and output variables with X_train,X_test,Y_train,Y_test position
X_train,X_test,Y_train,Y_test=train_test_split(independent, dependent, test_size=.30,random_state=0)

```

```python
# Applying simple linear regression algorithm by using sklearn
from sklearn.linear_model import LinearRegression
# assigning number of functions present in LinearRegression class into regressor variable
regressor=LinearRegression()
# assigning respected input and output variable by using fit method
regressor.fit(X_train,Y_train)
# once the above line executed then the weight and bias values are found
```

```python
weight=regressor.coef_
bias=regressor.intercept_
```

```python
# viewing weight value
weight
```

```python
# viewing bias value
bias
```

```python
# Once we get weight and bias then go with evaluation metrics to find the model is good or bad performance
Y_pred=regressor.predict(X_test)
```

```python
# 
from sklearn.metrics import r2_score
r_score=r2_score(Y_test,Y_pred)
```

```python
r_score
```

```python
# here we getting nearly to 1 ,so it is good mode
```

```python
# Save the model by using pickle
import pickle
filename="finalized_model_linear.sav"
```

```python
# Load the model by using pickle
pickle.dump(regressor,open(filename,'wb'))
```

```python
loaded_model=pickle.load(open("finalized_model_linear.sav",'rb'))
result=loaded_model.predict([[15]])
```
