```python
# pandas used read the data from database related tables
import pandas as pd
```

```python
# load the data from respected folder
dataset=pd.read_csv("Profit or Lose Data.csv")
```

```python
# viewing the data present in dataset
dataset
```

```python
# The above table showing categorical data in States column,so we need to convert into numerical data by using nominal method
# "One hot encoding"
dataset=pd.get_dummies(dataset,dtype=int,drop_first=True)
```

```python
dataset
```

```python
#viewing the columns present in the particular table 
dataset.columns
```

```python
#Assigning the inputs by using independent variable 
independent=dataset[['R&D Spend', 'Administration', 'Marketting Spend', 'Profit',
       'States_Florida', 'States_New York']]
```

```python
independent
```

```python
# Assigning the output by using dependent variable
dependent=dataset[["Profit"]]
```

```python
dependent
```

```python
# the following the importing particular function train_test_split from sklearn
from sklearn.model_selection import train_test_split
X_train,X_test,Y_train,Y_test=train_test_split(independent,dependent,test_size=0.30,random_state=0)
```

```python

from sklearn.linear_model import LinearRegression
# Loading all fuction present under linear regression to regressor variable
regressor=LinearRegression()
# by using fit method asiigning the input as X_train,output as Y_train
regressor.fit(X_train,Y_train)
```

```python
# After the above code will execute weight and bias values calculated 
weight=regressor.coef_
```

```python
# viewing the weight values 

weight
```

```python
bias=regressor.intercept_
```

```python
# viewing the bias values 

bias
```

```python
Y_pred=regressor.predict(X_test)
```

```python
Y_pred
```

```python
from sklearn.metrics import r2_score
```

```python
r_score=r2_score(Y_test,Y_pred)
```
